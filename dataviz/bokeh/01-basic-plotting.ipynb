```python
output_notebook()
```

```python
import bokeh.sampledata
bokeh.sampledata.download()
```

## Scatter

```python
# Scatter kullanılabilecek shape değerleri
# asterisk()
# circle()
# circle_cross()
# circle_x()
# cross()
# diamond()
# diamond_cross()
# hex()
# inverted_triangle()
# square()
# square_cross()
# square_x()
# triangle()
# x()

p = figure(plot_width=300, plot_height=300)

p.circle(x=[1,2,3], y=[9,8,0], size=10, line_color='navy', fill_color='red', fill_alpha=0.5)
p.square(x=[2,1,4], y=[3,5,7], size=10, line_color='navy', fill_color='blue', fill_alpha=0.5)
p.hex(x=[9,0,2], y=[1,4,8], size=10, line_color='navy', fill_color='orange', fill_alpha=0.5)

show(p)
```

```python
from bokeh.sampledata.autompg import autompg as df

df.head()
```

```python
p = figure(plot_width=300, plot_height=300, title="Scatter")

weight = df['weight']
yr = df['yr']

p.circle(x=weight, y=yr, size=10, line_color="navy", fill_alpha=0.5)

show(p)
```

## Line 

```python
p = figure(plot_width=300, plot_height=300, title="Line")

p.line([1,2,3,4,5], [3,4,1,5,9], line_width=2, line_dash='dashed')
p.circle([1,2,3,4,5], [3,4,1,5,9], color="navy", size=5) # Aynı anda da kullanılabiliyor figürler.

show(p)
```

```python
from bokeh.sampledata.glucose import data

data.head()
```

```python
timerange = data.loc['2010-10-01':'2010-10-03']

p = figure(x_axis_type="datetime")
p.xaxis.axis_label = 'Time'
p.yaxis.axis_label = 'Value'

p.line(timerange.index, timerange.glucose)

show(p)
```
