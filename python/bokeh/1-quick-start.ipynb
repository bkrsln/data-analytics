```python
from bokeh.plotting import figure, output_file, show
from bokeh.io import push_notebook, show, output_notebook
```

```python
x = [1, 2, 3, 4, 5, 9]
y = [6, 7, 2, 4, 3, 0]

output_file("1.html")

p = figure(title="Basit bir çizgi grafik", x_axis_label='x', y_axis_label='y')

p.line(x, y, legend_label="Çizgi değeri", line_width=2)

show(p)
```

```python
z = [1, 2, 3, 4, 5, 6]
y0 = [i**2 for i in z]
y1 = [10**i for i in z]
y2 = [10**(i**2) for i in z]

output_notebook()

p = figure(y_axis_type="log", y_range=[0.001, 10**11], title="log örneği", x_axis_label="sect", y_axis_label="part")

p.line(z, z, legend_label="z=z")
p.circle(z, z, legend_label="z=z")
p.line(z, y0, legend_label="y=z^2", line_color="red")
p.circle(z, y0, legend_label="y=z^2", fill_color="red", size=6)
p.line(z, y1, legend_label="y=10^z")
p.circle(z, y1, legend_label="y=10^z")
p.line(z, y2, legend_label="y=10^z^2", line_color="orange", line_dash="5 5")

show(p)
```

```python
import numpy as np
```

```python
N = 4000
x = np.random.random(size=N) * 100
y = np.random.random(size=N) * 100
radi = np.random.random(size=N) * 1.5
colors = ["#%02x%02x%02x" % (int(r), int(g), 150) for r, g in zip(50+2*x, 30+2*y)]
tools = "crosshair,pan,wheel_zoom,box_zoom,reset,box_select,lasso_select"

output_notebook()

p = figure(tools=tools, x_range=(0, 100), y_range=(0,100))
p.circle(x, y, radius=radi, fill_color=colors, fill_alpha=0.6, line_color=None)

show(p)
```
